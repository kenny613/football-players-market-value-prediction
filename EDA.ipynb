```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    df['Age'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1},
).set(xlabel='Age', ylabel='Count',title="Age");
plt.show()
sns.displot(data=df,x='Age', kind="kde", bw_adjust=2)
plt.show()
```

```python
sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.distplot(
    df['Market value'], norm_hist=False, kde=False, bins=20, hist_kws={"alpha": 1}
).set(xlabel='Market value', ylabel='Count',title="Market value");

plt.show()
sns.displot(data = df,x='Market value', kind="kde", bw_adjust=2)
plt.show()
sns.displot(data = df,x='Market value', kind="kde", bw_adjust=2,hue="Position",multiple="stack")
plt.show()
```

```python
df=pd.read_csv(r"dataset_v7.csv",index_col=[0])
f, axes = plt.subplots(3, 3,figsize = (21,21))
sns.set_style("whitegrid")

attr_lst = ['Age','Goals','Assists','Yellow Cards','Dispossessed','Pass Completion %','Successful Dribble %','Tackles Won','Market value']
for i,attr in enumerate(attr_lst):
    sns.boxplot(x = 'Position', y = attr_lst[i], data = df, ax=axes[int(i/3)][int(i%3)]).set(title="%s vs Position" %attr_lst[i])

plt.show()
```
